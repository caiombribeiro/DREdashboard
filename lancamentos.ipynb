```python
import pandas as pd


with open("janeiro_junho","r+",encoding="ISO-8859-1") as arquivo:
    linhas = arquivo.readlines()

    contador = 0
    linhas_final = []
    dados = ""
    ult_dia = ""

    for valor in linhas[1:-1]:

        valor = valor[9:]
        
        
             
        if valor[0] != "0":
            contador += 1

            dados += f"{valor[1:11]},"
            ult_dia = dados.split(",")[0]

        else:
            if len(dados) != 0:

                dados += f"{valor[3:7]},{valor[10:14]},{valor[14:27]}.{valor[27:29]},{valor[-108:-101]},{valor[36:155]}"
        
                linhas_final.append(dados)
                dados = ""

            else:

                contador += 1
                dados = f"{ult_dia},{valor[3:7]},{valor[10:14]},{valor[14:27]}.{valor[27:29]},{valor[-108:-101]},{valor[36:155]}"

                linhas_final.append(dados)
                dados = ""


```

```python
#Aqui eu farei um dicionrio usando as contas contbeis do DRE
contas = {}
with open("lista2.txt","r") as filtro:
    for linha in filtro:
        drop = linha.strip()
        partes = drop.split(" ")
        key = partes[0]
        # Verifica se a chave  um dgito
        if key.isdigit():  
            key = int(key)  # Converte para inteiro
            value = partes[-1]
            contas[key] = value

print(contas)
```

### Fato

```python

with open("teste.csv", "w",encoding="latin-1") as file:
    for linha in linhas_final:
        linha = linha.split(",")
        debito = f'{linha[0]},{linha[1]},-{linha[3]},{linha[4]},{linha[5]}'
        credito = f'{linha[0]},{linha[2]},{linha[3]},{linha[4]},{linha[5]}'

        if linha != linhas_final[-1]:
            file.writelines(f'{debito}\n{credito}\n')

        else:
            file.writelines(f'{debito}\n{credito}')
```

```python

#Aqui eu farei um dicionrio usando as contas contbeis do DRE
contas = {}
with open("lista2.txt","r") as filtro:
    for linha in filtro:
        drop = linha.strip()
        partes = drop.split(" ")
        key = partes[0]
        # Verifica se a chave  um dgito
        if key.isdigit():  
            key = int(key)  # Converte para inteiro
            value = partes[-1]
            contas[key] = value

print(contas)
```

```python
tabela = pd.read_csv("teste.csv", encoding="latin-1", header = None)

# Condio de filtro com base nas chaves presentes no contas.keys()
condicao =  (tabela[1].isin(contas.keys()))
# (tabela[1] != '0000') & (tabela[1] != 0) &

# Aplicando a condio para filtrar as linhas
tabela = tabela.loc[condicao].reset_index(drop=True)

tabela['CodContaContabil'] = tabela[1].map(contas)
#Renomear o cabealho
tabela = tabela.rename(columns={0:"Data",1:"Conta",2:"Valor",3:"Empresa",4:"Descricao"})
tabela = tabela[["Data","Conta","CodContaContabil","Valor","Empresa","Descricao"]]

tabela
```
